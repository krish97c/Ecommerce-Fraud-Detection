```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from google.colab import files
# Load dataset
data = pd.read_csv('/content/Fraud_Data_with_Countries.csv')

# Load IP address to country mapping CSV file
ip_to_country_mapping = pd.read_csv('/content/IpAddress_to_Country (1).csv')

# Function to map IP address to country
def ip_to_country(ip):
    ip_int = int(ip)
    country = ip_to_country_mapping[(ip_int >= ip_to_country_mapping['lower_bound_ip_address']) & (ip_int <= ip_to_country_mapping['upper_bound_ip_address'])]['country']
    return country.values[0] if not country.empty else 'Unknown'

# Apply the function to map IP addresses to countries
data['country'] = data['ip_address'].apply(ip_to_country)

# Feature Engineering for 'user_id'
data['user_id_length'] = data['user_id'].apply(lambda x: len(str(x)))
data['user_id_domain'] = data['user_id'].apply(lambda x: str(x).split('@')[-1])

# Feature Engineering for 'purchase_value'
data['purchase_value_bins'] = pd.cut(data['purchase_value'], bins=3, labels=['low', 'medium', 'high'])

# Encode categorical features
label_encoder = LabelEncoder()
data['user_id_domain_encoded'] = label_encoder.fit_transform(data['user_id_domain'])
data['country_encoded'] = label_encoder.fit_transform(data['country'])

# Save the modified dataset to a new CSV file
data.to_csv('modified_dataset.csv', index=False)


# Save the modified dataset to a new CSV file
data.to_csv('modified_dataset.csv', index=False)
# Download the CSV file
files.download('modified_dataset.csv')
```
