```python
import pandas as pd
import numpy as np
from google.colab import files
# Load dataset from a CSV file
file_path = '/content/modified_dataset.csv'  # Replace 'your_dataset.csv' with the path to your CSV file
original_data = pd.read_csv(file_path)

# Function to decrease dataset size using sampling
def decrease_dataset_size(data, sample_size):
    """
    data: pandas DataFrame, original dataset
    sample_size: int, desired size of the reduced dataset
    """
    if sample_size >= len(data):
        return data  # No need to decrease, return original data

    # Randomly sample 'sample_size' number of rows from the original dataset
    reduced_data = data.sample(n=sample_size, random_state=42)  # You can change the random_state if needed

    return reduced_data

# Set the desired sample size
desired_sample_size = 30000  # Same as original dataset size

# Decrease dataset size
reduced_data = decrease_dataset_size(original_data, desired_sample_size)

# Display reduced dataset
print(reduced_data.head())

# Save reduced dataset to a CSV file
reduced_data.to_csv('reduced_dataset22.csv', index=False)  # Change 'reduced_dataset.csv' to your desired filename
# Download the CSV file
files.download('reduced_dataset22.csv')
```
